```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# 1. Cargar datos (Asegúrate de haber montado el Drive antes)
ruta = "/content/drive/MyDrive/Colab Notebooks/AI Project/"
train_df = pd.read_csv(ruta + 'train_18_24_pca_final.csv')
test_df = pd.read_csv(ruta + 'test_25_pca_final.csv')

features = [f'PC{i+1}' for i in range(6)]
X_train_full = train_df[features]
y_train_full = train_df['Target']
X_test = test_df[features]
y_test = test_df['Target']

# 2. Crear una muestra reducida para SVM y KNN (Plan B)
# 30,000 es un número ideal para que terminen en < 5 minutos sin perder mucha precisión
X_train_sample, y_train_sample = resample(X_train_full, y_train_full, n_samples=30000, random_state=42)

# 3. Diccionario de modelos
# Los rápidos van con el set FULL, los lentos con el SAMPLE
modelos = {
    "Regresión Logística": (LogisticRegression(max_iter=1000), X_train_full, y_train_full),
    "SVM": (SVC(kernel='rbf', probability=True), X_train_sample, y_train_sample),
    "KNN": (KNeighborsClassifier(n_neighbors=5), X_train_sample, y_train_sample),
    "Random Forest": (RandomForestClassifier(n_estimators=100), X_train_full, y_train_full),
    "XGBoost": (XGBClassifier(eval_metric='logloss'), X_train_full, y_train_full)
}

resultados_stats = []
modelos_entrenados = {}

print("Iniciando entrenamiento híbrido (Full + Subsampling)...")

for nombre, (modelo, X_ptr, y_ptr) in modelos.items():
    print(f"--- Entrenando {nombre}... ---")
    modelo.fit(X_ptr, y_ptr)

    # Evaluar siempre contra el TEST FULL (2025)
    y_pred = modelo.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    resultados_stats.append({
        "Modelo": nombre,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-Score": f1
    })
    modelos_entrenados[nombre] = modelo
    print(f"✓ {nombre} completado.")

# 4. Generar Tabla Comparativa
df_res = pd.DataFrame(resultados_stats)
print("\n" + "="*40)
print("RESULTADOS FINALES (TAREA 4)")
print("="*40)
print(df_res.sort_values(by="F1-Score", ascending=False).to_string(index=False))

# 5. Graficar Matrices de Confusión (Para tu documento)
plt.figure(figsize=(20, 15))
for i, nombre in enumerate(modelos_entrenados.keys(), 1):
    plt.subplot(3, 2, i)
    y_pred = modelos_entrenados[nombre].predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
    plt.title(f'Matriz de Confusión: {nombre}')
    plt.ylabel('Real')
    plt.xlabel('Predicho')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

# 1. Montar Drive y Cargar Datos
drive.mount('/content/drive')
ruta = "/content/drive/MyDrive/Colab Notebooks/AI Project/"

def cargar_csv_robusto(path):
    # Intenta detectar si el separador es , o ;
    try:
        df = pd.read_csv(path, sep=',')
        if len(df.columns) < 2:
            df = pd.read_csv(path, sep=';')
        return df
    except:
        return None

train_df = cargar_csv_robusto(ruta + 'train_18_24_pca_final.csv')
test_df = cargar_csv_robusto(ruta + 'test_25_pca_final.csv')

# 2. Definir Features (PC1 a PC14)
# Dinámicamente buscamos todas las columnas que empiecen con 'PC'
features = [c for c in train_df.columns if c.startswith('PC')]
X_train_full = train_df[features]
y_train_full = train_df['Target']
X_test = test_df[features]
y_test = test_df['Target']

print(f"Entrenando con {len(features)} componentes: {features}")
print(f"Dataset: {X_train_full.shape[0]} filas entrenamiento | {X_test.shape[0]} filas prueba")

# 3. Subsampling para modelos lentos (SVM y KNN)
# Usamos 30,000 para que termine en ~10 min con 14 variables
X_train_sample, y_train_sample = resample(X_train_full, y_train_full, n_samples=30000, random_state=42)

# 4. Configuración de Modelos (Tarea 4)
modelos = {
    "Regresión Logística": (LogisticRegression(max_iter=1000), X_train_full, y_train_full),
    "SVM": (SVC(kernel='rbf', probability=True), X_train_sample, y_train_sample),
    "KNN": (KNeighborsClassifier(n_neighbors=5), X_train_sample, y_train_sample),
    "Random Forest": (RandomForestClassifier(n_estimators=100, random_state=42), X_train_full, y_train_full),
    "XGBoost": (XGBClassifier(eval_metric='logloss', random_state=42), X_train_full, y_train_full)
}

resultados = []
print("\nIniciando entrenamiento de los 5 modelos...")

for nombre, (modelo, X_ptr, y_ptr) in modelos.items():
    print(f"Procesando {nombre}...")
    modelo.fit(X_ptr, y_ptr)
    y_pred = modelo.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    resultados.append({"Modelo": nombre, "Accuracy": acc, "F1-Score": f1})

    # Imprimir reporte rápido
    print(f"   > F1-Score: {f1:.4f}")

# 5. Visualización Final (Tabla y Gráficas)
df_res = pd.DataFrame(resultados).sort_values(by="F1-Score", ascending=False)
print("\n" + "="*40)
print("COMPARATIVA FINAL SEST ALERT SYSTEM")
print("="*40)
print(df_res.to_string(index=False))

# Matriz de Confusión del mejor modelo
mejor_modelo_nombre = df_res.iloc[0]['Modelo']
plt.figure(figsize=(8,6))
y_pred_best = modelos[mejor_modelo_nombre][0].predict(X_test)
sns.heatmap(confusion_matrix(y_test, y_pred_best), annot=True, fmt='d', cmap='Blues')
plt.title(f'Mejor Modelo: {mejor_modelo_nombre} (PC1-PC14)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, roc_auc_score

# 1. Montar Drive y Cargar Datos Nuevos
drive.mount('/content/drive')
ruta = "/content/drive/MyDrive/Colab Notebooks/AI Project/"

def cargar_csv(nombre):
    return pd.read_csv(ruta + nombre)

# Cargamos los datasets generados con SMOTE y 15 componentes
train_df = cargar_csv('train_preprocesado_balanceado.csv')
test_df = cargar_csv('test_preprocesado_2025.csv')

# 2. Definir Features (PC1 a PC15)
features = [f'PC{i+1}' for i in range(15)]
X_train_full = train_df[features]
y_train_full = train_df['Target']

# En el test, separamos las columnas de ID para el reporte final
X_test = test_df[features]
y_test = test_df['Target']
ids_test = test_df[['codigo', 'id_prov', 'id_cant', 'amie']]

print(f"Dataset Balanceado: {X_train_full.shape[0]} filas | Features: {len(features)}")

# 3. Muestreo para modelos pesados (SVM/KNN)
# Usamos el set balanceado para que la muestra sea 50/50 real
X_train_sample, y_train_sample = resample(X_train_full, y_train_full, n_samples=25000, random_state=42, stratify=y_train_full)

# 4. Configuración de Modelos (Sincronizado con Tarea 4)
modelos = {
    "Regresión Logística": (LogisticRegression(solver='lbfgs', class_weight='balanced', max_iter=1000), X_train_full, y_train_full),
    "SVM": (SVC(kernel='rbf', C=1.0, probability=True, random_state=42), X_train_sample, y_train_sample),
    "KNN": (KNeighborsClassifier(n_neighbors=5, weights='distance'), X_train_sample, y_train_sample),
    "Random Forest": (RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42), X_train_full, y_train_full),
    "XGBoost": (XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, eval_metric='logloss', random_state=42), X_train_full, y_train_full)
}

resultados = []
print("\nIniciando entrenamiento híbrido...")

for nombre, (modelo, X_ptr, y_ptr) in modelos.items():
    print(f"Entrenando {nombre}...")
    modelo.fit(X_ptr, y_ptr)

    y_pred = modelo.predict(X_test)
    y_prob = modelo.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    resultados.append({"Modelo": nombre, "Accuracy": acc, "F1-Score": f1, "AUC-ROC": auc})
    print(f"   > F1: {f1:.4f} | AUC: {auc:.4f}")

# 5. Resultados Finales
df_res = pd.DataFrame(resultados).sort_values(by="AUC-ROC", ascending=False)
print("\n" + "="*50)
print("RANKING DE MODELOS (BASADO EN AUC-ROC)") [cite: 342]
print("="*50)
print(df_res.to_string(index=False))

# Guardar predicciones del mejor modelo para el Mapa de Calor
mejor_mod_nombre = df_res.iloc[0]['Modelo']
mejor_modelo = modelos[mejor_mod_nombre][0]
test_df['Prob_Riesgo'] = mejor_modelo.predict_proba(X_test)[:, 1]
test_df['Pred_Riesgo'] = mejor_modelo.predict(X_test)
test_df.to_csv(ruta + 'predicciones_finales_sest.csv', index=False)

print(f"\nArchivo 'predicciones_finales_sest.csv' guardado con el modelo: {mejor_mod_nombre}")
```
